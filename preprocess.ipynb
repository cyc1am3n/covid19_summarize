```python
covid_metadata_dir = './data/covid_19_metadata.csv'
covid_meta = pd.read_csv(covid_metadata_dir, encoding='latin')
```

```python
covid_meta.head()
```

```python
covid_meta = covid_meta[['Title', 'Abstract', 'DOI', 'Authors']]
covid_meta = covid_meta.dropna(axis=0).reset_index(drop=True)
print(covid_meta.shape)
```

```python
covid_meta.head()
```

```python
import re
doi_url = re.compile("https://doi.org/[0-9.]*/.*")
doi = re.compile("[0-9.]*/.*")
```

```python
print(doi_url.match("https://doi.org/10.1016/j.msard.2020.102182"))
print(doi_url.match("10.1097/aog.0000000000003950"))
print(doi.match("10.1097/aog.0000000000003950"))
```

```python
from langdetect import detect

for i in range(covid_meta.shape[0]):
    title, abstract = covid_meta.loc[i, 'Title'], covid_meta.loc[i, 'Abstract']
    if detect(title) == 'en' and detect(abstract) == 'en':
        row = covid_meta.iloc[i:i+1]
        if doi.match(row.loc[i, 'DOI']):
            row.loc[i, 'DOI'] = "https://doi.org/" + row.loc[i, 'DOI']
        elif doi_url.match(row.loc[i, 'DOI']):
            pass
        else:
            continue

        if len(abstract) < 25: # No available Abstract
            continue
        elif abstract[:10].lower() == 'background':
            if abstract[10:12] == ' :':
                abstract = abstract[13:]
            elif abstract[10] == ':':
                abstract = abstract[12:]
            elif abstract[10] == ' ':
                abstract = abstract[11:]
            else:
                abstract = abstract[10:]
        elif abstract[:8].lower() == 'abstract':
            abstract = abstract[9:]
        elif abstract[:7].lower() == 'summary':
            abstract = abstract[8:]
        elif abstract[:3] == '<p>':
            abstract = abstract[3:-4]
        row.loc[i, 'Abstract'] = abstract
        
        if i == 0:
            english_covid = row
        else:
            english_covid = pd.concat([english_covid, row])
```
